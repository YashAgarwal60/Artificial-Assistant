```python
engine=pyttsx3.init('sapi5')
voices=engine.getProperty('voices')
engine.setProperty('voice',voices[1].id)
```

```python
def speak(audio):
    engine.say(audio)
    engine.runAndWait()
```

```python
def takecommand():
    r=sr.Recognizer()
    with sr.Microphone() as source:
        print('\nListening...')
        r.pause_threshold=1
        r.adjust_for_ambient_noise(source, duration=1) 
        audio=r.listen(source, phrase_time_limit=8) 
    try:
         print('Recognizing...')
         query=r.recognize_google(audio,language='en-in')
         print(f"user said: {query}\n")
    except sr.UnknownValueError:
        print("Sorry, I can't hear you! Please repeat again... or check your internet connection")
        speak("Sorry, I can't hear you! Please repeat again... or check your internet connection")
        return 'None'
    return query
```
